Data visualisation:-
By the help of data  visualisation , we can easily get the insights using dashboards , for data visualisation matplotlib.pyplot is to be imported.
By using matplotlib library we have to find out:-
1) which is the highest sellng category.
2) which is the highest  selling sub-category of product.
3) which paymentmode opted by customers most frequently.
4) what is the relationship between profit and sales?
5)  which paymentmode fetched the highest sales? 


1) Finding the most selling category of product:-

```python
plt.bar(df['Category'],df['Amount'],color='red')
plt.title('Category v/s Amount')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

2) Finding the most selling sub-category of product:-

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/EXCEL/Details.csv")
plt.barh(df['Sub-Category'],df['Amount'])
plt.title('sub-category v/s amount')
plt.xlabel('Amount')
plt.ylabel('Sub-category')
plt.show()
```

3) the paymentmode opted by customers most frequently:-

```python
plt.bar(df['PaymentMode'],df['Amount'],color='gold')
plt.title('paymentmode v/s amount')
plt.xlabel('Amount')
plt.ylabel('paymentmode')
plt.show()
```

4) finding the relationship between amount and profit:-

```python

plt.scatter(df['Profit'],df['Amount'])
plt.xlabel('Profit',fontsize=12)
plt.ylabel('Amount',fontsize=12)
plt.grid(True,linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```
